# Calculating the 'score'

```python
df = pd.read_csv('zomato_restaurants.csv')
df.head()
```

```python
average_rating = df['rating'].mean()
print(average_rating)
```

```python
# Minimum number of votes required to be in the charts
min_votes = df['votes'].quantile(0.50)
print(min_votes)
```

```python
# Restaurants that qualify for the chart
q_rest = df.copy().loc[df['votes'] >= min_votes]
q_rest.shape
```

```python
# Computing weighted rating of each restaurant
def weighted_rating(df, m=min_votes, C=average_rating):
    v = df['votes']
    R = df['rating']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
q_rest['score'] = round(q_rest.apply(weighted_rating, axis=1), 3)
```

```python
q_rest = q_rest.sort_values('score', ascending=False)
df_copy = q_rest[['rest_index', 'rest_name', 'locality', 'cost_for_two', 'rating', 'votes', 'score', 'cuisines']]
```

```python
df_copy.to_csv('zomato_qualified_restaurants.csv', index=False)
```

# Content-Based Recommendations

```python
df = pd.read_csv('zomato_restaurants.csv')
df.head(10)
```

```python
print(df.loc[df['rest_name']=='Punjab Grill']['rating'])
```

```python
df['cuisines'] = df['cuisines'].fillna("").astype('str')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(df['cuisines'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim[:4, :4]
```

```python
# Build a 1-dimensional array with movie titles
titles = df['rest_name']
indices = pd.Series(df.index, index=df['rest_name'])

# Function that get movie recommendations based on the cosine similarity score of movie genres
def genre_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    rest_indices = [i[0] for i in sim_scores]
    final = titles.iloc[rest_indices].head()
    return final.values
```

```python
print(genre_recommendations("Taaza Thindi"))
```

# Collaborative Filtering

```python
df = pd.read_csv('zomato_reviews - FINAL.csv')
df.head()
```

```python
userRatings = df.pivot_table(index=['cust_name'], columns=['rest_name'],values='cust_rating')
userRatings.head(252)
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head(10)
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix.head()
```

```python
userRatings.to_csv('userRatings.csv', index=True)
corrMatrix.to_csv('corrMatrix.csv', index=True)
```
